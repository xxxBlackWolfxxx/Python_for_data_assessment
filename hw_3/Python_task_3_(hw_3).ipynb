Определение самого популярного мужского и женского имени на корабле (применение mode)

```python
def extract_name(name):
  #Случай для женского имени, которое находится в скобках
    if '(' in name:
        return name.split('(')[1].split(')')[0].split(' ')[0].strip(' "\'')
  #Случай для мужского имени вне скобок
    else:
        return name.split('. ')[1].split(' ')[0].strip(' "\'')

titanic_df['FirstName'] = titanic_df['Name'].apply(extract_name)

male_names = titanic_df[titanic_df['Sex'] == 'male']['FirstName']
female_names = titanic_df[titanic_df['Sex'] == 'female']['FirstName']

print(f'Самое популярное мужское имя на корабле: {male_names.mode()[0]}')
print(f'Самое популярное женское имя на корабле: {female_names.mode()[0]}')
```

Самые популярные мужские и женские имена по классам

```python
for pclass in [1, 2, 3]:
    print(f"\nКласс {pclass}:")
    class_df = titanic_df[titanic_df['Pclass'] == pclass]
    male_name = class_df[class_df['Sex'] == 'male']['FirstName'].mode()
    female_name = class_df[class_df['Sex'] == 'female']['FirstName'].mode()
    print(f"Мужское: {male_name[0]}")
    print(f"Женское: {female_name[0]}")
```

Вывод количества n-местных кабин на корабле

```python
cabin_counts = titanic_df['Cabin'].value_counts().dropna()
for n in range(1, cabin_counts.max() + 1):
  if n == 1:
    print(f"Кабин с {n} человеком: {(cabin_counts == n).sum()}")
  else:
    print(f"Кабин с {n} людьми: {(cabin_counts == n).sum()}")
```
