Импорт библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from google.colab import files
```

Загрузка датасета

```python
uploaded = files.upload()
df = pd.read_csv("train.csv")
```

Количество выживших

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived')
plt.title('Распределение выживших пассажиров')
plt.xlabel('Выжил (1 - да, 0 - нет)')
plt.ylabel('Число людей')
plt.show()
```

Количество пассажиров по классам

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass')
plt.title('Распределение пассажиров по классам')
plt.xlabel('Класс')
plt.ylabel('Число людей')
plt.show()
```

Количество пассажиров в зависимости от возраста

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='Age')
plt.title('Распределение возраста пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Число людей')
plt.show()
```

Количество пассажиров по полу

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex')
plt.title('Распределение пассажиров по полу')
plt.xlabel('Пол')
plt.ylabel('Число людей')
plt.show()
```

Распределение признака Parch

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Parch')
plt.title('Признак Parch')
plt.xlabel('Количество родителей/детей') #один из переводов этого признака
plt.ylabel('Число людей')
plt.show()
```

Boxplot диаграмма для признака Age

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, y='Age')
plt.title('Boxplot распределения возраста пассажиров')
plt.ylabel('Возраст')
plt.show()
```

Диаграмма показывает, что медианный возраст пассажиров около 28 лет. Большинство пассажиров (межквартильный диапазон) находятся в возрасте от 20 до 38 лет. Значения выбросы - возраст пассажиров старше 65 лет.

Pie chart для переменной Survived

```python
plt.figure(figsize=(6, 6))
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Non-survivors', 'Survivors'])
plt.title('Доля выживших пассажиров')
plt.ylabel('')
plt.show()
```

Pie chart для переменной Pclass

```python
plt.figure(figsize=(6, 6))
df['Pclass'].value_counts().plot.pie(autopct='%1.1f%%', labels=['3 класс', '1 класс', '2 класс'])
plt.title('Распределение пассажиров по классам')
plt.ylabel('')
plt.show()
```

Pairplot для всех числовых переменных

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
numerical_cols = numerical_cols.drop("PassengerId") #PassengerId не является полезным для анализа признаком
sns.pairplot(df[numerical_cols])
plt.suptitle('Pairplot числовых переменных', y=1.02)
plt.show()
```

Sunburst plot для признаков класса и пола

```python
fig = px.sunburst(
    df,
    path=['Pclass', 'Sex'],
    title='Распределение пассажиров по классам и полу'
)
fig.show()
```
